## Modular arithmetic:

$$\frac{A}{B} = Q\ \text{remainder}\ R$$

In R, *%/%* computes the integer division (i.e., Q), *%%* computes the remainder (i.e., R).


```r
# integer devision '%/%'
10 %/% 3
13 %/% 5
-5 %/% 3
2 %/% 0
```

```r
# modulo '%%'
10 %% 3
13 %% 5
-5 %% 3
2 %% 0
```

# Matrix and Array

A matrix is a two-dimensional array that contains elements of the same type, arranged in rows and columns. [R Programming: Zero to Pro - Matrix](https://r02pro.github.io/matrix.html)

A array can be viewd as an generalization of vector and matrix to a higher dimensional space, and still contains elements of the same type. [R Programming: Zero to Pro - Array](https://r02pro.github.io/array.html)

```r
# Create a 3X4 matrix, with number 1 to 12
mat1 = matrix(1:12, nrow = 3, ncol = 4)
mat1
```

```r
## Let's get more information about this matrix
# the dimension
dim(mat1)

# the number of rows
nrow(mat1)

# the number of columns
ncol(mat1)

# the class
class(mat1)

# the internal storage type
typeof(mat1)

# attribute
attributes(mat1)
```

# Graphics

In R, the basic graphic function is plot().

```r
# First, let's generate some vectors that we can use to explore plot()
xval <- 1:20
yval_1 <- 20:1
yval_2 <- runif(20, min = 5, max = 15)
```

```r
# plot(x), single variable
plot(xval)
```

```r
# bivariate plot of x and y
plot_1 = plot(x = xval, y = yval_1)

# If we would like to change the type of the plot or customize the plot
# use 'type =' to specify the type of plot.
# axes = TRUE/FALSE: draw or do not draw the axes and the box
# xlim =, ylim =: specify the lower and upper limits of the axes
# xlab =, ylab =: annotate the axes
# main =: main title
# find more options with help()

plot_1 = plot(x = xval, y = yval_1, type = 'l', xlim = c(-1,21), ylim = c(-1,21), col = 'red')
# if you would like to add a line or points to an already existing plot
points(x = xval, y = yval_2, pch = 15, col = 'blue')
```
